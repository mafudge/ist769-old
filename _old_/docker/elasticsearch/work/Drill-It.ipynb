```python
def drill_sql(sql_select, echo = False):
    from sqlalchemy import create_engine
    from sqlalchemy.sql import text 
    import pandas as pd
    engine = create_engine("drill+sadrill://drill:8047/dfs?use_ssl=False", echo = echo)
    statement = text(sql_select)
    with engine.connect() as conn:    
        df = pd.read_sql(statement, conn)
        return df
```
