```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .master("local") \
    .appName('jupyter-pyspark') \
    .getOrCreate()
sc = spark.sparkContext
sc.setLogLevel("ERROR")
```

```python
df = spark.read.format("org.neo4j.spark.DataSource") \
  .option("url", "bolt://neo4j:7687") \
  .option("labels", "Person") \
  .load()
df.show()
```

```python
cipher_ql = "MATCH (person:Person {name: 'Tom Hanks'})-[:ACTED_IN]->(movie:Movie) RETURN person.name, movie.title, movie.released, movie.tagline"
df = spark.read.format("org.neo4j.spark.DataSource") \
  .option("url", "bolt://neo4j:7687") \
  .option("query",cipher_ql) \
  .load()
df.toPandas()
```
